```python
import seaborn as sns
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
np.mean(df['total_bill'])
```

```python
from scipy import stats
```

```python
# for mode, we need library scipy
stats.mode(df['tip'])
```

```python
# [0] is used to select the first mode, as .mode() can return multiple if
# there's a tie. This code uses the .mode() method which is built into pandas
# Series objects.
df['sex'].mode()[0]
```
